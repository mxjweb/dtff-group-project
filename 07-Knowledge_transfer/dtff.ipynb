```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(np.random.randn(100, 4), columns=list('ABCD'))
```

```python
df.plot()
plt.title("Random series")
plt.show()
```

```python
df_cumsum = df.cumsum(axis=0)
```

```python
df_cumsum.plot()
plt.title("Cumulative series")
```
